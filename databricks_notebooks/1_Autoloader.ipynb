### AutoLoader - Incremental Load

```python
%sql
create schema netflix_catalog.net_schema
```

```python
from pyspark.sql.functions import * 
```

```python
checkpoint_path = 'abfss://silver@netflixstorageaccount.dfs.core.windows.net/checkpoint'
```

```python
df = spark.readStream\
    .format('cloudFiles')\
    .option('cloudFiles.format', 'csv')\
    .option('cloudFiles.schemaLocation', checkpoint_path)\
    .load('abfss://raw@netflixstorageaccount.dfs.core.windows.net')
```

```python
display(
    df,
    # checkpointLocation = "abfss://silver@netflixstorageaccount.dfs.core.windows.net/checkpoint"
)
```

```python
df.writeStream\
  .option('checkpointLocation', checkpoint_path)\
  .trigger(processingTime='10 seconds')\
  .start('abfss://bronze@netflixstorageaccount.dfs.core.windows.net/netflix_titles')
```
